```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap='plasma')
plt.title("Missing Values Heatmap", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='orange')
plt.title("Distribution of Passenger Ages", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title("Survival Count by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.grid(axis='y')
plt.tight_layout()
plt.show()
print("Women had significantly higher survival rates than men")
```

```python
sns.countplot(data=df, x='Survived', hue='Sex', palette='cool')
plt.title("Survival Count by Gender")
plt.legend(title='Gender', loc='upper right')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), kde=True, bins=30, color='skyblue')
plt.title("Age Distribution of Titanic Passengers", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived', palette='viridis')
plt.title("Survival Count Across Passenger Classes")
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class'])
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("train.csv")

# Calculate survival rate by passenger class
pclass_survival = df.groupby('Pclass')['Survived'].mean().reset_index()

# Plot
plt.figure(figsize=(7, 5))
sns.barplot(x='Pclass', y='Survived', data=pclass_survival, palette='coolwarm')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class'])
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
